```python
import pandas as pd
import numpy as np
import math
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('/Users/scrysam/Downloads/hotels.csv')
#df
df.describe()
```

```python
#df['distribution_channel']
#df['hotel']
df = df.drop(columns=["distribution_channel"])
df = df.drop(columns=["country"])
df = df.drop(columns=["arrival_date"])
grouped = df.groupby(df.hotel) 
df_res = grouped.get_group("Resort Hotel")
df_res = df_res.drop(columns=["hotel"])
df_res
```

```python
#Error. You have given 'city hotel' all small letters, but actually it is 'City Hotel'
grouped = df.groupby(df.hotel) 
df_cty = grouped.get_group("City Hotel") 
df_cty = df_cty.drop(columns=["hotel"])
df_cty
```

```python
X1 = df_res.drop(columns=['is_canceled'])
X1
```

```python
X2 = df_cty.drop(columns=['is_canceled'])
X2
```

```python
y1 = df_res['is_canceled']
y1
```

```python
y2 = df_cty['is_canceled']
y2
```

```python
x = X1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X1 = pd.DataFrame(x_scaled)

X1
```

```python
x = X2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X2 = pd.DataFrame(x_scaled)

X2
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)

```

```python
model = DecisionTreeClassifier()
model.fit(X1_train,y1_train)
predictions = model.predict(X1_test)

score = accuracy_score(y1_test, predictions)
score
```

```python
model = DecisionTreeClassifier()
model.fit(X2_train,y2_train)
predictions = model.predict(X2_test)

score = accuracy_score(y2_test, predictions)
score
```
