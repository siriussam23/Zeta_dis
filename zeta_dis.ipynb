```python
import pandas as pd
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('/Users/scrysam/Downloads/2021-01-04_zeta-disease_dsi-take-home-challenge.csv')
df.describe()
```

```python
mean_arr = df.describe(include='all').loc['mean']
mean_arr
#mean_arr["bmi"]
```

```python
df["bmi"] = df["bmi"].replace([0],mean_arr["bmi"])
```

```python
df["blood_pressure"] = df["blood_pressure"].replace([0],mean_arr["blood_pressure"])
df["insulin_test"] = df["insulin_test"].replace([0],mean_arr["insulin_test"])
df["cardio_stress_test"] = df["cardio_stress_test"].replace([0],mean_arr["cardio_stress_test"])

df.describe()
```

```python
X = df.drop(columns=['zeta_disease'])
X
```

```python
y = df['zeta_disease']
y
```

```python
x = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X = pd.DataFrame(x_scaled)

X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)

score = accuracy_score(y_test, predictions)
score
```
